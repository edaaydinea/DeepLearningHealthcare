### 1. If you are given a dataset of 100 X-ray images, would you still use CNN models? If so, which architecture would you try? And why?

- Yes, I would still use a **CNN** model for 100 X-ray images. However, given the relatively small dataset size, I would try **transfer learning** using a pre-trained CNN architecture such as **ResNet-50**, **DenseNet**, or **VGG**. These architectures, pre-trained on large datasets like ImageNet, have learned useful features and can be fine-tuned on smaller datasets. Transfer learning helps prevent overfitting when training with limited data by leveraging learned features from a larger dataset.

### 3. Which of the following is NOT true about convolutional neural networks (CNN)?

(d) **CNN usually has much more parameters than fully connected neural networks**.

This is incorrect. CNNs are designed to have fewer parameters than fully connected neural networks because of **local connections** and **weight sharing** mechanisms, making them more efficient in image processing tasks.

### 4. Given input sequence = [1, 2, 3, 4, 5, 6] and a filter [1, 2, 1], what is the output of the convolution operation with stride 2 and no padding?

- The convolution process with stride 2 involves applying the filter on every second element without overlapping.

Steps:
- (1×1 + 2×2 + 3×1) = 1 + 4 + 3 = 8
- (3×1 + 4×2 + 5×1) = 3 + 8 + 5 = 16
- (5×1 + 6×2) = 5 + 12 = 17

So, the output is **[8, 16, 17]**.

### 5. What is the size of the output feature map given 6x6 input and two filters of 4x4 with stride 1 and no padding?

- The formula for output size is:

$$
\text{Output size} = \left(\frac{n - f}{\text{stride}} + 1\right)
$$

Where:
- $n = 6$ (input size)
- $f = 4$ (filter size)
- $\text{stride} = 1$

$$
\text{Output size} = \left(\frac{6 - 4}{1} + 1\right) = 3
$$

Thus, the output feature map size is **3x3** for each filter. Since there are two filters, the feature map size will be **3x3x2**.

### 6. What is the number of parameters for a convolutional layer given a 6x6 input and two filters of 4x4 with stride 1 and no padding?

- For each filter, the number of parameters is $4 \times 4 = 16$. Since there are two filters, the total number of parameters is:

$$
16 \times 2 = 32
$$

If bias terms are included, the number of parameters becomes **34** (32 parameters for weights + 2 for biases).

### 7. What is the output of 1D max pooling over input [1, 2, 3, 2, 3, 5] with a filter of size 2 and stride 2?

- Max pooling with filter size 2 and stride 2:

Steps:
- Max(1, 2) = 2
- Max(3, 2) = 3
- Max(3, 5) = 5

The output is **[2, 3, 5]**.

### 8. What is NOT true about AlexNet?

(b) **AlexNet uses larger filters of 11x11 which is computationally more efficient than subsequent CNN models such as VGG and ResNet**.

This is false. Larger filters are **less** computationally efficient than the smaller filters used in later models like VGG and ResNet, which introduced smaller filter sizes (e.g., 3x3) for better efficiency and performance.

### 9. What is NOT true about CNN architectures?

This is incorrect. Inception Net uses **1x1 filters** for dimensionality reduction, but they do not necessarily ensure the dimensions across different paths are the same.
